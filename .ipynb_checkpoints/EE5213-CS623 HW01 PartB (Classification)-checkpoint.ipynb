## Preparing the Data

Currently face_images and noface_images are stored as a sequence (list) of images and each image is a 3D matrix with dimension (100, 100, 3). As discussed in class we need to convert each image to a single column vector and create the input matrix $X$ for the multi-input regression problem:

- We prepare a new 2D numpy array <b>X</b> with dimensions ($n$, $m$); where $n$ is the number of pixels in the image and $m$ is total data instances available. We flatten out each image and store it as a single column in the array $X$. 

- Labels are stored in a vector $\vec{y}$ of dimensions (1, $m$). Where each entry represents the label of corresponding image in $X$.

- Finally We shuffle the dataset set so that we can seperate training data and test data. 

Flattening out images so all the pixel is stored as a single vector of size (image_size x image_size x 3)

````python
image_size = 100

number_of_data_instances = len(noface_images)+len(face_images)
X = np.zeros((image_size*image_size*3, number_of_data_instances), dtype=np.int)

for i in range(len(face_images)):
    X[:, i] = face_images[i].flatten()
for i in range(len(noface_images)):
    X[:, i+len(face_images)] = noface_images[i].flatten()
    
print(X.shape)
````

Preparing the labels of both classes.

````python
# y lables for images (1 for Face-class, 0 for NoFace-Class)

noface_labels = np.zeros((1, len(noface_images)))
face_labels = np.ones((1, len(face_images)))
Y = np.column_stack((face_labels, noface_labels))

print(Y.shape)
````

Shuffling both $X$ and $\vec{y}$.

````python
# shuffle the data instances

permuations = np.random.permutation(number_of_data_instances)
X = np.take(X, permuations, axis=1)
Y = np.take(Y, permuations, axis=1)

print(X.shape)
print(Y.shape)
````

### Data Normalization
Pixels in an image file are stored as integers in the range 0-255. However to improve gradient descent calculation, it is common to normalize the pixels to floating point values in the 0-1 range. We accomplish that simply by diving each pixel by 255.

````python
X = X/255.
````

### Loss Function (Graded)

- Compute predicted output vector: $\vec{y'}$ = $\vec{w}$.<b>X</b> + b

  Hint: Use np.dot for vectorized implementation.

- The loss function $L_{mse}$ can then be calculated from observed output vector $\vec{y}$ and predicted output vector $\vec{y'}$ as:

    Hint: Use np.sum for vectorized implementation.

\begin{equation}
 L_{mse} = \frac{1}{m} \sum_{i=1}^{m} (y'^{(i)}-y^{(i)})^2
\end{equation}
  
   

````python
def compute_loss(w, b, X, Y):
    
    ########### WRITE YOU CODE BELOW ######
    m = X.shape[1]
    cost = 0
    cost = (1/m)*(np.sum(np.square(Y - ( (w@X) + b) )))
    
    ########### END OF CODE ###############
    
    return cost
````

You can test you loss function now.

### Calculate Gradient (Graded)
- Compute predicted output vector: $\vec{y'}$ = $\vec{w}$.<b>X</b> + b
- For each weight $j = 1, 2, ..., n$ calculate $\frac{dL}{dw_j}$ as $dw$:
\begin{equation}
 \frac{dL}{dw_{j}} = \Delta w_j = \frac{1}{m} \sum_{i=1}^{m} (y'^{(i)}-y^{(i)}).x^{(i)}
\end{equation}

As a final result return a single vector $\vec{dw}$ where each entry contain the derrivative wrt to that weight. 

- For bias calculate $\frac{dL}{db}$ as $db$:
\begin{equation}
 \frac{dL}{db} = \Delta b =  \sum_{i=1}^{m} \frac{1}{m}(y'^{(i)}-y^{(i)})
\end{equation}

Hint: The obvious implementation is loop based, which is extremely slow. You can use np.multipy and np.sum to compute all the gradient effeciently.

````python
def calculate_gradient(X, Y, weights, bias):
    
    m = X.shape[1]
    
    ########### WRITE YOU CODE BELOW ######
    
    dw = None
    db = None
    
    dw = (1/m * (weights@X + bias - Y)@ X.T)[0]
    db =  1/m * np.sum(((weights@X + bias) - Y))
    
    ########### END OF CODE ###############
    
    return dw, db
````

### Predict Function (Graded)

- Compute predicted output vector: $\vec{y'}$ = $\vec{w}$.<b>X</b> + b
- If $\vec{y'^{(i)}} \ge 0.5$ then replace by 1, otherwise replace by $0$.

````python
def predict(w, b, X):
    
    ########### WRITE YOU CODE BELOW ######
    
    y_prime = w@X + b
    y_prime = np.where(y_prime>=0.5,1.,0.)
    
    ########### END OF CODE ###############
    
    return y_prime
````

### Training your model (Graded)
This will be your training loop. For every iteraion you will do the following:
- Calcualte the $\vec{dw}$ and $db$ using compute_gradient function.
- Update the weights for $j=1, 2, ... , n$ and bias $b$ using:
\begin{equation}
 w_j = w_j - (lr * dw_j)
\end{equation}
\begin{equation}
 b = b - (lr * db) 
\end{equation}

    HINT: Again, use vectorized implementation instead of naive loop-based implementation to update all the weights and bias.

- Compute the cost for the updated weights and store it in the history for analysis later. 

````python
def train_classifier(weights, bias, X, Y, iterations, learning_rate, log_frequency):
    
    cost_history  = []
    
    for i in range(iterations):
        
        ########### WRITE YOU CODE BELOW ######
        
        # computing the gradient
        dw,db = calculate_gradient(X, Y, weights, bias)

        # updating the parameters
        weights -=learning_rate * dw
        bias -= learning_rate*db
        
        # calculate the cost for analysis later
        cost = compute_loss(weights, bias, X, Y)
        
        ########### END OF CODE ###############
        
        
        if i % log_frequency == 0:
            cost_history.append(cost)
            print ("Cost after iteration %i: %f" %(i, cost))
        
    return weights, bias, cost_history
````

### Hyperparamters, Weight Initializaiton and Starting Training (Graded)

If cost is not diverging, the learning rate is high and need to reduced. If cost function is diverging but very slowy, you need to bring up the learning rate. <b>In the end you should be able to reduce the loss as low as 0.05 (or less). </b>
- (Graded) <b>Epochs: </b>Numer of gradient descent steps to carry out. You can also try the number in the range [200, 5000] and see how it effects training. First start with fewer number of iterations, if training is working perfectly and loss is decreasing on each iteration (constult the epoch vs loss graph below), you can train for more iterations. 
- Initialization of weights and bias.
- Call to the function train_classifier with the desired inputs, outputs and parameters defined already. 

````python
########### WRITE YOU CODE BELOW ######

# learning rate
learning_rate = 0.0001

# training epochs
iterations = 10000

########### END OF CODE ###############

# initializing the weights and bias
weights = np.zeros((1, X_train.shape[0]))
bias = 0

# boolean for whether to print the logs
print_cost = True

# frequency with which to print the logs
log_frequency = iterations/25
    
print("Learning Rate: ", learning_rate)
print()

# recording the time before starting learning
start_time = time.time()

# training the classifier
w, b, cost_history = train_classifier(weights, bias, X_train, y_train, iterations, learning_rate, log_frequency)

# measuring the end time and computing mintues/secs
end_time = time.time()
time_taken = end_time - start_time #converting it to minutes
minutes = int(time_taken/60)
seconds = int(time_taken%60)

print("\nTime taken to train the model: {min_key: .02f} min/s, {sec_key: 0.2f} secs".format(min_key=minutes, sec_key=seconds))
````
